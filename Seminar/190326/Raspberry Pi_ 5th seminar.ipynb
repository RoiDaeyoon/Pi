## 2. Jupyter Notebook 설치 및 TensorFlow를 이용한 병렬 계산
 
 
 ### 2.1 pip를 통한 Jupyter 설치
Raspberry에서 파이썬 라이브러리를 설치하는 방법은 저번 시간에 사용했던 pip를 이용하여 설치가 가능하다.  
pip란, 파이썬으로 작성된 소프트웨어 패키지를 설치하고 관리할 때 유용한 패키지 관리 시스템이다.

- _sudo pip3 install --upgrade pip_ : pip를 최신버전으로 업그레이드
- _sudo pip3 install jupyter_ : pip를 이용해 jupyter를 다운로드


설치 완료까지의 시간은 약 10분 정도 소요되었다.

이후, 명령어로 _jupyter notebook_ 을 입력하면 아래와 같은 창이 뜨면서 jupyter notebook을 실행한다.



_jupyter notebook_ 은 단지 실행을 위한 코드일뿐, 연산을 실행하기 위해서는 Raspberry Pi의 IP주소를 통해 접속을 해야한다.  

 ### 2.2 IP 주소를 통한 jupyter 실행
IP 주소를 통한 jupyter 실행 명령어는 다음과 같다.  
_jupyter-notebook --ip="ip주소" --no-browser_  
  
명령어를 입력하여 실행이 되었다면 맨 마지막에 URL 주소가 나오게 되는데, 이 주소를 통해 인터넷에서 접속하면 jupyter를 실행할 수 있다.  


 ### 2.3 TensorFlow
마찬가지 방법으로 _sudo pip3 install tensorflow_ 명령어를 이용해 tensorflow를 설치할 수 있다.  

__Clustering을 통한 병렬연산__을 하는 방법 뿐만 아니라 __Tensorflow Package를 이용__하여 병렬연산이 가능하다는 것을 알게 되었다.  
같은 네트워크 서버를 사용하고 있다면 GPU를 사용해 병렬 연산이 가능하다.  
이를 위해선 __CudaToolkit 7.0과 CUDNN 6.5 v2__ 라는 프로그램을 사용해야 한다.




__쿠다(CUDA)__는 Compute Unified Device Architecture의 약자로,  
그래픽 처리 장치(GPU)에서 수행하는 (병렬 처리) 알고리즘을 산업 표준 언어를 사용하여 작성할 수 있도록 하는 GPGPU 기술입니다. _ 위키백과 발췌

쿠다를 이용하면 CUP에서 순차적으로 처리되는 많은 양의 데이터를 GPU의 병렬처리 작업을 함으로서 구글의 슈퍼컴퓨터에서나 가능했던 작업이 일반 데스크탑에서 가능하게 됩니다.
(Hadoop이나 Spark가 독립된 컴퓨터 즉, CPU기반의 병렬 처리인 반면 CUDA는 GPU기반의 병렬처리방식)

__CUDA의 데이터 흐름__
- 그래픽 카드 메모리 공간을 할당한다.
- PC의 입력 데이터를 그래픽 카드의 메모리로 복사한다.
- 강력한 GPU 성능을 이용하여 병렬 처리한다.
- 처리된 결과를 그래픽 카드의 메모리에서 PC의 메모리로 복사한다.
